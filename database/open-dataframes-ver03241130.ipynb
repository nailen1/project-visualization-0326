```python
def getPricePerPy_category(nameCategory):
    if df_total[df_total['category']==nameCategory]['area_py'].sum() != 0:
        return int(df_total[df_total['category']==nameCategory]['price_estimate'].sum()/df_total[df_total['category']==nameCategory]['area_py'].sum())
    else:
        return 0
```
